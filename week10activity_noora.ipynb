# 3. Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and comment on your observations

```python
# Finding the data types of 'Hardness','Sulfate' and 'Organic Carbon'
df[['Hardness','Sulfate','Organic_carbon']].dtypes
```

It is clear that the data types of the variables 'Hardness', 'Sulfate' and 'Organic Carbon' are float datatype.

```python
# Plotting the histogram of 'Hardness', 'Sulfate' and 'Organic Carbon' therby analyzing the nature of distribution
plt.figure(figsize=(20,8))
plt.subplot(1,3,1)
plt.hist(df['Hardness'],color='red')
plt.title('Histogram of Hardness')
plt.grid()
plt.subplot(1,3,2)
plt.hist(df['Sulfate'],color='green')
plt.title('Histogram of Sulfate')
plt.grid()
plt.subplot(1,3,3)
plt.hist(df['Organic_carbon'],color='purple')
plt.title('Histogram of Organic_carbon')
plt.grid()
```

From the graphs , we can say that all the three features 'Hardness','Sulfate' and 'Organic carbon' are normally distributed

```python
# Plotting the boxplots of 'Hardness', 'Sulfate' and 'Organic Carbon' therby checking the presence of outliers
plt.figure(figsize=(20,8))
plt.subplot(1,3,1)
plt.boxplot(df['Hardness'])
plt.title('Boxplot of Hardness')
plt.grid()
plt.subplot(1,3,2)
plt.boxplot(df['Sulfate'])
plt.title('Boxplot of Sulfate')
plt.grid()
plt.subplot(1,3,3)
plt.boxplot(df['Organic_carbon'])
plt.title('Boxplot of Organic_carbon')
plt.grid()
```

From the above plots, it is seen that all the three features 'Hardness', 'Sulfate' and 'Organic Carbon 'contain outliers

# 4. Carry out correlation analysis on the dataset and find the correlation factor between


# a. ph and Chloramines

# b.ph and Conductivity

# What do we infer from this correlation value?

```python
# Correlation Analysis of entire data set
corrmatrix1=df.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix1,vmin=-.2,vmax=.8,annot=True,linewidth=0.2)
```

```python
# Finding the correlation factor between 'ph' and 'Chloramines'
df1=df[['ph','Chloramines']]
corrmatrix2=df1.corr()
plt.subplots(figsize=(8,4))
sns.heatmap(corrmatrix2,vmin=-.2,vmax=.8,annot=True,linewidth=0.2,cmap='Blues')
```

Here the correlation factor between ph and Chloromines is -0.025 which is approximately equal to zero .That means there is no 
significant correlation between the variables

```python
# Finding the correlation factor between 'ph' and 'Conductivity'
df2=df[['ph','Conductivity']]
corrmatrix3=df2.corr()
plt.subplots(figsize=(8,4))
sns.heatmap(corrmatrix3,vmin=-.2,vmax=.8,annot=True,linewidth=0.2,cmap='gray')
```

From the plot the correlation factor between ph and Conductivity is 0.014 which is also approximately equal to zero.  Hence there is 
no correlation between the variables.
